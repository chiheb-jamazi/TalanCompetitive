```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from bs4 import BeautifulSoup
from time import sleep
import time
import pandas as pd
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from hdfs import InsecureClient
```

```python
ratings =[]
working_dates =[]
review_headlines =[]
review_links =[]
author_job_titles =[]
review_pros =[]
review_cons =[]
def get_glassdoor_posts (soup):
    containers = soup.find_all(class_='gdReview')
    for container in containers :
        try :
            rating = container.find('span')
            rating = rating.get_text()
            ratings.append(rating)
    #         print(ratings)
        except :
            ratings.append('none')
        ###Description######
        try :
            working_date = container.find(class_='pt-xsm pt-md-0 css-1qxtz39 eg4psks0')
            working_dates.append(working_date.get_text())
    #         print(working_dates)
        except :
            working_dates.append('none')
        ###Likes######
        try:
            review_headline = container.find(class_='reviewLink')
            review_headlines.append(review_headline.get_text())
    #         print(current_position)
        except :
            review_headlines.append('none')
        ###comments_count######
        try :
            review_link = container.find(class_='mb-xxsm mt-0 css-93svrw el6ke055')
            review_links.append(review_link.a['href'])
    #         print(review_links)
        except:
            review_links.append('none')
        ###shares_count ######  
        try :
            author_job_title = container.find(class_='authorJobTitle middle common__EiReviewDetailsStyle__newGrey')
            author_job_titles.append(author_job_title.get_text())
    #         print(author_job_titles)
        except :
            author_job_titles.append('none')
        ###images link ###### 
        try :
            review_pro = container.find(class_='mt-0 mb-0 pb v2__EIReviewDetailsV2__bodyColor v2__EIReviewDetailsV2__lineHeightLarge v2__EIReviewDetailsV2__isCollapsed')
            review_pros.append(review_pro.get_text())
    #         print(review_pros)
        except :
            review_pros.append('none')
        try :
            review_con = container.find_all(class_='mt-0 mb-0 pb v2__EIReviewDetailsV2__bodyColor v2__EIReviewDetailsV2__lineHeightLarge v2__EIReviewDetailsV2__isCollapsed')
            review_cons.append(review_con[1].get_text())
    #         print(review_cons)
        except :
            review_cons.append('none')
```

```python
post_dates =[]
CandidateSubtexts =[]
Inteview_titles =[]
ApplicationDetails =[]
InterviewProcess =[]
InterviewQuestions =[]
InterviewRatings =[]
def get_glassdoor_interviews (soup):
    containers = soup.find_all(class_='mt-0 mb-0 my-md-std p-std gd-ui-module css-cup1a5 ec4dwm00')
    for container in containers :
        try : 

            post_date = container.find(class_='mt-0 mb-xxsm d-flex justify-content-between css-13r90be e1lscvyf1')
            post_dates.append(post_date.get_text())
            post_dates
        except :
            post_dates.append('none')

        try :

            CandidateSubtext = container.find(class_='mt-0 mb css-13r90be e1lscvyf1')
            CandidateSubtexts.append(CandidateSubtext.get_text())
            CandidateSubtexts
        except :
            CandidateSubtexts.append('none')

        try :

            inteview_title = container.find(class_='mt-0 mb-xxsm css-93svrw el6ke055')
            Inteview_titles.append(inteview_title.get_text())
            Inteview_titles
        except :
            Inteview_titles.append('none')

        try : 

            ApplicationDetail = container.find(class_='mt-xsm mb-std')
            ApplicationDetails.append(ApplicationDetail.get_text())
            ApplicationDetails
        except :
            ApplicationDetails.append('none')

        try :

            InterviewProces = container.find(class_='css-w00cnv mt-xsm mb-std')
            InterviewProcess.append(InterviewProces.get_text())
            InterviewProcess
        except :
            InterviewProcess.append('none')

        try : 

            InterviewQuestion = container.find(class_='d-inline-block mb-sm')
            InterviewQuestions.append(InterviewQuestion.get_text())
            InterviewQuestions
        except :
            InterviewQuestions.append('none')
        try :

            InterviewRating = container.find_all(class_='d-block d-sm-inline-block mr mb-xsm mb-sm-0')
            InterviewRatings.append([i.get_text() for i in InterviewRating])
            InterviewRatings
        except:
            InterviewRatings.append('none')
```

```python
def scrolling_function():
    start = time.time()
    initialScroll = 0
    finalScroll = 10
  
    while True:
        driver.execute_script(f"window.scrollTo({initialScroll},{finalScroll})")
        initialScroll = finalScroll
        finalScroll += 10
        end = time.time()
        if round(end - start) > 3:
            break 
```

```python
url = "https://www.glassdoor.com/profile/login_input.htm?userOriginHook=HEADER_SIGNIN_LINK"
# driver = webdriver.Chrome(r'C:\\Users\\jchih\\\Documents\\selenium\\chromedriver.exe')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.maximize_window()
driver.get(url)
sleep(3)
connect_to_glassdoor ()
sleep(3)
urls = ['https://www.glassdoor.com/Reviews/Talan-Reviews-E485967.htm',
        'https://www.glassdoor.com/Interview/Talan-Interview-Questions-E485967.htm']
for url in urls:
    driver.get(url)
    if url == 'https://www.glassdoor.com/Reviews/Talan-Reviews-E485967.htm' :
        for i in range(5) :
            scrolling_function ()
            sleep(.5)
            company_page = driver.page_source
            soup =BeautifulSoup(company_page, 'lxml')
            get_glassdoor_posts (soup)
            sleep(1)
            job_buttons = driver.find_element_by_css_selector("#MainContent > div > div:nth-child(1) > div.d-flex.flex-column.align-items-top > div > div.pageContainer > button.nextButton.css-1hq9k8.e13qs2071")
            job_buttons.click()  
            sleep(3)

        Reviews_Data = pd.DataFrame(
                        {'ratings': ratings,
                         'working_dates': working_dates,
                         'review_headlines': review_headlines,       
                         "review_links":review_links,
                         "author_job_titles": author_job_titles,
                         "review_pros":review_pros,
                         "review_cons":review_cons,
                        })
    else:
        for i in range(6) :
            scrolling_function ()
            sleep(.5)
            containers = soup.find_all(class_='mt-0 mb-0 my-md-std p-std gd-ui-module css-cup1a5 ec4dwm00')
            company_page = driver.page_source
            soup =BeautifulSoup(company_page, 'lxml')
            get_glassdoor_interviews (soup)
            sleep(.5)
            next_button = driver.find_element_by_css_selector("#EmployerInterviews > div.d-flex.flex-column.align-items-top.mt-std.mt-md-0.mb-std > div > div.pageContainer > button.nextButton.css-1hq9k8.e13qs2071")
            next_button.click()
            sleep(3)

        Interviews_Data = pd.DataFrame(
                        {'post_dates': post_dates,
                         'CandidateSubtexts': CandidateSubtexts,
                         'Inteview_titles': Inteview_titles,       
                         "ApplicationDetails":ApplicationDetails,
                         "InterviewProcess": InterviewProcess,
                         "InterviewQuestions":InterviewQuestions,
                         "InterviewRatings":InterviewRatings,
                        })
```
